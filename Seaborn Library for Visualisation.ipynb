```python
# If I want to turn off confidence Interval off bu putting ci = false
sns.relplot(x = 'timepoint', y = 'signal' , kind = 'line',data = fmri, ci = False)
```

```python
#Sinplot and variation in the graph w.r.t. axes
def sinplot(flip =1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*0.5)*(7-i)*flip)
```
